```python
import scipy.stats as stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Carbonate_Chemistry = "C:/Users/Dina/Downloads/group3-project-main/group3-project-main/mergingdatasetFINAL2.xlsx"
data= pd.ExcelFile(Carbonate_Chemistry)

sheet_names = data.sheet_names

print(sheet_names)

ph_control = pd.read_excel(data, sheet_name = 'pH control')

ph_elevated_pc02 = pd.read_excel(data, sheet_name ='pH Elevated PCO2')

water_quality = pd.read_excel(data, sheet_name ='Water Quality')

mcav_control = pd.read_excel(data, sheet_name ='Mcav Control')

mcav_elevated_PC02 = pd.read_excel(data, sheet_name ='Mcav Elevated PCO2')

ofav_control = pd.read_excel(data, sheet_name ='Ofav Control')

ofav_evelavted_p_c02 = pd.read_excel(data, sheet_name = 'Ofav Elevated P CO2')

pcli_control = pd.read_excel(data, sheet_name = "Pcli Control")

pcli_evelvated_pc02 = pd.read_excel(data, sheet_name = 'Pcli Elevated PCO2')

mcav_control.head()
```

**pH**

```python
ph_merged = pd.concat([ph_control, ph_elevated_pc02], axis = 0).reset_index(drop=True)
ph_merged.head()
```

```python
palette = {"Control": "yellow", "Elevated ": "red"}
sns.boxplot(data = ph_merged, x = "Week", y = "pH", hue = "Dataset", palette = palette).set_title("pH, Control vs Elevated pCO2")
```

**Mcav**

```python
mcav_merged = pd.concat([mcav_control, mcav_elevated_PC02], axis = 0).reset_index(drop=True)
mcav_merged.head()
```

```python
palette = {"Control": "yellow", "Elevated ": "red"}
sns.boxplot(data = mcav_merged, x = "Day Measured", y = "Surface Area Density (g/mm2)", hue = "Dataset", palette = palette).set_title("Mcav Surface Area Density, Control vs Elevated pCO2")
plt.show()
```

```python
palette = {"Control": "yellow", "Elevated ": "red"}
fig, ax = plt.subplots()
mcav_subset = mcav_merged.loc[mcav_merged["Day Measured"] > 0, :]
sns.boxplot(data = mcav_subset, x = "Day Measured", y = "Calcification rate (mgCaCO3/cm2/day)", hue = "Dataset", palette = palette).set_title("Mcav Calcification Rate, Control vs Elevated pCO2")
plt.show()
```

```python
palette = {"Control": "yellow", "Elevated ": "red"}
fig, ax = plt.subplots()
mcav_subset = mcav_merged.loc[mcav_merged["Day Measured"] > 0, :]
sns.boxplot(data = mcav_subset, x = "Day Measured", y = "Tissue growth (mm2/day)", hue = "Dataset", palette = palette).set_title("Mcav Tissue Growth, Control vs Elevated pCO2")
# ax.set_xlim(28, 84)
plt.show()
```

```python
palette = {"Control": "yellow", "Elevated ": "red"}
fig, ax = plt.subplots()
mcav_subset = mcav_merged.loc[mcav_merged["Day Measured"] > 0, :]
sns.boxplot(data = mcav_subset, x = "Day Measured", y = "% change Surface Area Density", hue = "Dataset", palette = palette).set_title("Mcav % Change Surface Area Density, Control vs Elevated pCO2")
# ax.set_xlim(28, 84)
plt.show()
```
